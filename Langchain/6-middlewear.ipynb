- Transforming prompts, tool selection, and output formatting.

- Applying rate limits, guardrails, and PII detection.

```python
# Summarization Middleware Function
def compress_history(history, model, keep_last=3):
    """Auto-compress history when too long"""
    if len(history) <= keep_last:
        return history
    
    # Summarize old messages
    old = history[:-keep_last]
    summary = model.invoke([
        *old,
        HumanMessage(content="Summarize in 2 sentences.")
    ]).content
    
    # Return summary + recent messages
    recent = history[-keep_last:]
    return [SystemMessage(content=f"Past: {summary}")] + recent

# Test
long_chat = [
    HumanMessage(content="I'm Sarah."),
    AIMessage(content="Hi!"),
    HumanMessage(content="Love coding."),
    AIMessage(content="Nice!"),
    HumanMessage(content="Use Python."),
    AIMessage(content="Great!"),
    HumanMessage(content="Live in London."),
]

print(f"Before: {len(long_chat)}")
compressed = compress_history(long_chat, model)
print(f"After: {len(compressed)}")
print(f"Summary: {compressed[0].content[:60]}...")
```

```python
# Simple Summarization Example
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

# Setup model
model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

# Long conversation
chat_history = [
    HumanMessage(content="Hi, I'm Alex from New York."),
    AIMessage(content="Hello Alex! Nice to meet you."),
    HumanMessage(content="I'm learning Python."),
    AIMessage(content="Great choice! Python is awesome."),
    HumanMessage(content="I work as a teacher."),
    AIMessage(content="That's a wonderful profession."),
]

print(f"Messages before: {len(chat_history)}")

# Summarize
summary_prompt = [
    *chat_history,
    HumanMessage(content="Summarize this in one sentence.")
]
summary = model.invoke(summary_prompt).content
print(f"Summary: {summary}")

# Compressed history
compressed = [
    SystemMessage(content=f"Context: {summary}"),
    HumanMessage(content="Where do I live?")
]

answer = model.invoke(compressed).content
print(f"Answer: {answer}")
```
