```python
from pathlib import Path

base_dir = Path("/home/moeen/projects/AgenticAI/RAG/data/text_files")
base_dir.mkdir(parents=True, exist_ok=True)

python_file = base_dir / "python_programming.txt"
machine_learning_file = base_dir / "machine_learning.txt"

python_content = """Python Programming

Python is a high-level programming language known for its simplicity and readability.
It is widely used for web development, automation, data analysis, artificial intelligence,
and scientific computing. Python supports multiple programming paradigms, including
procedural, object-oriented, and functional programming.

Some important features of Python include easy syntax, dynamic typing, extensive libraries,
and a large community of developers. Beginners often prefer Python because it allows them
to write clear and maintainable code with fewer lines compared to many other languages.

Python is commonly used with frameworks such as Flask and Django for web applications.
It is also used with libraries such as NumPy, Pandas, and Matplotlib for data science.
In machine learning, Python is very popular because of tools like Scikit-learn, TensorFlow,
and PyTorch.

A simple Python program can print text, perform calculations, and process user input.
Variables, loops, functions, and classes are fundamental concepts in Python programming.
Learning Python helps students and developers build practical solutions for real-world problems.
"""

machine_learning_content = """Machine Learning

Machine learning is a branch of artificial intelligence that enables computers to learn
patterns from data and make predictions or decisions without being explicitly programmed.
It is used in applications such as recommendation systems, spam detection, image recognition,
and natural language processing.

Machine learning models are usually trained on large datasets. During training, the model
learns relationships between input features and target outputs. After training, the model
is tested on new data to measure its accuracy and generalization ability.

There are three main types of machine learning: supervised learning, unsupervised learning,
and reinforcement learning. Supervised learning uses labeled data, while unsupervised
learning finds hidden patterns in unlabeled data. Reinforcement learning teaches an agent
to make decisions through rewards and penalties.

Common machine learning algorithms include linear regression, logistic regression, decision
trees, random forests, support vector machines, and neural networks. Deep learning is a
specialized area of machine learning that uses multi-layer neural networks for complex tasks.

Machine learning has become very important in modern technology because it helps automate
intelligent decision-making and improve the efficiency of many systems.
"""

python_file.write_text(python_content, encoding="utf-8")
machine_learning_file.write_text(machine_learning_content, encoding="utf-8")

print(f"Saved Python content to: {python_file}")
print(f"Saved machine learning content to: {machine_learning_file}")

from langchain_core.documents import Document

documents = [
    Document(
        page_content=python_content,
        metadata={"source": str(python_file), "title": "Python Programming", "type": "info"},
    ),
    Document(
        page_content=machine_learning_content,
        metadata={"source": str(machine_learning_file), "title": "Machine Learning", "type": "info"},
    ),
]

print(f"Prepared {len(documents)} documents for further processing.")
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader(
    "/home/moeen/projects/AgenticAI/RAG/data/text_files/machine_learning.txt",
    encoding="utf-8"
)

documents = loader.load()

print(documents)
```

```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader

# Load all .txt files from the directory
dir_loader = DirectoryLoader(
    "/home/moeen/projects/AgenticAI/RAG/data/text_files",
    glob="**/*.txt",          # Load all text files recursively
    loader_cls=TextLoader,    # Use TextLoader for each file
    loader_kwargs={"encoding": "utf-8"},
    show_progress=True
)

documents = dir_loader.load()

print(f"Number of documents loaded: {len(documents)}")

for doc in documents:
    print(doc.metadata["source"])
    print(doc.page_content[:100])  # Print first 100 characters
    print("-" * 50)
```
