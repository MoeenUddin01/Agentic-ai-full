```python
print("cell 0")
import os 
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("OPENROUTER_API_KEY")
if api_key:
    os.environ["OPENROUTER_API_KEY"] = api_key
```

```python
print("cell 1")
import os 
from typing import Literal
from langchain_openai import ChatOpenAI

from deepagents import create_deep_agent
from tavily import TavilyClient

tavily_api_key = os.getenv("TAVILY_API_KEY")
if tavily_api_key:
    tavily = TavilyClient(api_key=tavily_api_key)

llm = ChatOpenAI(
    model="gpt-4o-mini",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)
```

```python
from typing import Literal


def internet_search(
    query: str,
    max_results: int = 5,
    topic: Literal["general", "news", "finance"] = "general",
    include_raw_content: bool = False
):
    """
    Run a web search using Tavily API
    """
    return tavily.search(
        query,
        max_results=max_results,
        include_raw_content=include_raw_content,
        topic=topic,
    )


research_subagent = {
    "name": "research_agent",
    "description": "used to research more in depth questions",
    "system_prompt": "you are a great researcher",
    "tools": [internet_search],
    "model": llm
}

subagents = [research_subagent]

agent = create_deep_agent(
    model=llm,
    subagents=subagents,
)

# Display the agent graph structure
print("Agent Graph Structure:")
print(agent.get_graph().draw_mermaid())
```

```python
# Display the agent graph as a Mermaid diagram
from IPython.display import Image, display

# Get the graph as a Mermaid diagram
mermaid_diagram = agent.get_graph().draw_mermaid()
print("Agent Graph (Mermaid Diagram):")
print(mermaid_diagram)
```

```python
# Invoke the main agent with a task that will be delegated to the research subagent
print("=== Invoking Agent ===")
response = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": "Research the latest developments in artificial intelligence and provide a summary."
        }
    ]
})

print("\n=== Agent Response ===")
print(response)
```
