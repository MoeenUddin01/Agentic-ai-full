Pydantic

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

# Initialize your Groq model (specify model and provider separately)
model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

# Test the model
response = model.invoke("please tell what is the ai")
print(response.content)
```

```python
from pydantic import BaseModel,Field
class Movie(BaseModel):
    title: str = Field(description="Title of the movie")
    director: str = Field(description="Director of the movie")
    year: int = Field(description="Year of the movie")
    rating: float = Field(description="Rating of the movie")
```

```python
model_with_structured_output=model.with_structured_output(Movie)
model_with_structured_output
```

```python
model.invoke("provide details about the movie inception")
```

```python
output=model_with_structured_output.invoke("provide details about the movie inception")
print(output)
```

TypedDict


```python
#TypedDict provides a simpler alternative using python build in typing ideal when you don't need runtime validation 
```

```python
from typing_extensions import TypedDict,Annotated

class MovieDict(TypedDict):
    """Movie information"""
    title: Annotated[str,"Title of the movie"]
    director: Annotated[str,"Director of the movie"]
    year: Annotated[int,"Year of the movie"]
    rating: Annotated[float,"Rating of the movie"]

model_with_structured_output = model.with_structured_output(MovieDict)
response = model_with_structured_output.invoke("provide details about the movie inception")
print(response)
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    name: str = Field(description="Name of the person")
    phone: str = Field(description="Phone number of the person")
    email: str = Field(description="Email address of the person")

# Use with_structured_output instead of create_agent
model_with_contact = model.with_structured_output(ContactInfo)

result = model_with_contact.invoke(
    "Extract contact info from this text: John Doe, 123-456-7890, john.doe@example.com"
)
print(result)
```
