# Deep Agent Workflow

This notebook demonstrates creating both a **DeepAgent** and a **Basic LangChain Agent** using Groq LLM and Tavily Search.

## 1. Setup

Load environment variables and configure API keys.

```python
print("Cell 0")
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
```

## 2. Internet Search Tool

Define a web search tool using Tavily API.

```python
print("Cell 1")
#Tool internet search
from typing import Literal
from langchain_core.tools import tool
from tavily import TavilyClient

tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

@tool
def web_search(query: str, max_results: int = 5, topic: Literal["general", "news", "finance"] = "general", include_raw_content: bool = False):
    """
    Search the web using Tavily API
    """
    response = tavily_client.search(query=query, max_results=max_results, topic=topic, include_raw_content=include_raw_content)
    return response
```

## 3. Initialize Groq Model

Set up the Llama 3.3 70B model via Groq.

```python
print("Cell 2")
from langchain.chat_models import init_chat_model

groq_model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")
print(groq_model)
```

## 4. Create a Deep Agent

Use the `deepagents` library to create a research agent with web search capability.

```python
print("Cell 3")
##create a deep agent
from deepagents import create_deep_agent

deep_agent = create_deep_agent(
    tools=[web_search],
    model=groq_model,
    system_prompt="You are a researcher"
)
deep_agent
```

## 5. Create a Basic Agent

Use standard LangChain to create a simple agent.

```python
print("Cell 4")
#normal agent
#basic agent
from langchain.agents import create_agent

simple_agent = create_agent(
    model=groq_model
)
```
