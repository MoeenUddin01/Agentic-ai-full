# Agent Comparison: Basic Agent vs DeepAgent

This notebook demonstrates and compares a basic LangChain agent with a DeepAgent, both using Groq and Tavily.

## Setup

Load environment variables and import necessary libraries.

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage
from langchain_tavily import TavilySearch
from IPython.display import Image, display
from graphviz import Digraph

# Load environment variables
load_dotenv()

# Verify API keys are loaded
groq_api_key = os.getenv("GROQ_API_KEY")
tavily_api_key = os.getenv("TAVILY_API_KEY")

print(f"GROQ_API_KEY loaded: {bool(groq_api_key)}")
print(f"TAVILY_API_KEY loaded: {bool(tavily_api_key)}")
print("Setup complete!")
```

```python
# --- BASIC AGENT SECTION ---

## Basic Agent with LangChain

#Create a basic agent using standard LangChain with Groq and Tavily.
```

```python
# Initialize Basic LangChain Agent with Groq
basic_llm = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

print("✓ Basic LangChain Agent initialized")
print(f"Model: llama-3.3-70b-versatile")
```

### Basic Agent Architecture Diagram

```python
# Create Basic Agent Architecture Diagram
dot = Digraph(comment='Basic Agent Architecture')
dot.attr(rankdir='LR')

# User input
dot.node('User', 'User Query')

# Basic Agent components
dot.node('LLM', 'Groq LLM\n(llama-3.3-70b-versatile)')
dot.node('Response', 'Response')

# Edges
dot.edge('User', 'LLM')
dot.edge('LLM', 'Response')

# Render and display
print("=== BASIC AGENT ARCHITECTURE ===")
display(Image(dot.render(format='png')))
```
