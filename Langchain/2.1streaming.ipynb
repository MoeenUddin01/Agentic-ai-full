```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("groq:llama-3.3-70b-versatile")
model
```

```python
model.invoke("write me a 200 words paragrap of artifical intelligence")
```

```python
#now using the streaming
for chunk in model.stream("write me a 200 words paragrap of artifical intelligence"):
    print(chunk.content, end="", flush=True)
```

```python
#now using the streaming
for chunk in model.stream("write me a 200 words paragrap of artifical intelligence"):
    print(chunk.text, end="", flush=True)
```

```python
# Cell 6: Streaming with proper paragraph formatting

prompt = (
    "Write a 200-word essay about artificial intelligence. "
    "Make sure to explicitly split your response into two distinct paragraphs."
)

print("--- STARTING STREAM ---\n")

# Use chunk.content to safely catch all characters, including newlines (\n)
for chunk in model.stream(prompt):
    print(chunk.content, end="", flush=True)

print("\n\n--- STREAM FINISHED ---")
```

BATCH 

```python
prompts = [
    "Write a short paragraph about the history of Artificial Intelligence.",
    "Write a short paragraph about the future of Quantum Computing.",
    "Write a short paragraph about the impact of Biotechnology."
]

print("--- SENDING BATCH REQUESTS ---")

# .batch() takes a list and returns a list of responses
responses = model.batch(prompts)

print("--- ALL RESPONSES RECEIVED ---\n")

# Iterate through the results and display them cleanly
for i, response in enumerate(responses):
    print(f"=== Topic {i+1} ===")
    print(response.content)
    print("\n" + "-"*40 + "\n")
```

```python
# using the max_concurrency parameter to limit the number of concurrent requests

# 1. Your list of questions
prompts = [
    "Explain machine learning in one sentence.",
    "Explain deep learning in one sentence.",
    "Explain NLP in one sentence.",
    "Explain computer vision in one sentence."
]

# 2. Tell LangChain to process exactly 2 items at a time
config = {"max_concurrency": 2}

# 3. Run the batch (LangChain handles the splitting automatically)
responses = model.batch(prompts, config=config)

# 4. Print results using a simple loop index
for i in range(len(prompts)):
    print(f"\nQuestion: {prompts[i]}")
    print(f"Answer: {responses[i].content}")
```
