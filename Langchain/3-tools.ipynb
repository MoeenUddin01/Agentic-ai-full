```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
import langchain
```

```python
# Make sure your Groq API key is set
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# 1. Correct import path (lowercase 'l')
from langchain.chat_models import init_chat_model

# 2. Initialize your Groq model (use the format that worked: provider:model)
model = init_chat_model("groq:llama-3.3-70b-versatile")

# 3. Invoke it directly (no extra response imports needed)
response = model.invoke("What is the weather in Dubai?")

print(response.content)
```

```python
from langchain_core.tools import tool

@tool
def get_weather(location: str) -> str:
    """Get the current weather for a specific location."""
    # This is a mock implementation. In production, you'd call a real weather API
    weather_data = {
        "sf": "Sunny, 72°F",
        "san francisco": "Sunny, 72°F",
        "nyc": "Cloudy, 65°F",
        "new york": "Cloudy, 65°F",
        "dubai": "Hot, 95°F",
        "london": "Rainy, 55°F",
    }
    return weather_data.get(location.lower(), f"Weather data not available for {location}")

# Bind the tool to the model
tools = [get_weather]
model_with_tools = model.bind_tools(tools)
```

```python
# Step 1: Model generates tool calls
messages = [{"role": "user", "content": "what is the weather in sf?"}]
ai_mess = model_with_tools.invoke(messages)
messages.append(ai_mess)

# Step 2: Execute tool calls and collect the results
for tool_call in ai_mess.tool_calls:
    # Execute the tool with the generated arguments
    tools_result = get_weather.invoke(tool_call)
    messages.append(tools_result)

# Step 3: Pass results back to model for final response
final_response = model_with_tools.invoke(messages)
print(final_response.content)
```

```python
response = model_with_tools.invoke("What is the weather in Tokyo?")
print(response)
for tool_call in response.tool_calls:
    print("Tool Call:", tool_call["name"])
    print(f"Args: {tool_call['args']} ")
```
