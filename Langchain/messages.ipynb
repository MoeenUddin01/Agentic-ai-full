DETAILED SYSTEM PROMPT

```python
# Cell 20: LLM Agent with a Detailed System Prompt (No Tools)
from langchain_core.messages import SystemMessage, HumanMessage
from langchain.chat_models import init_chat_model

# 1. Initialize your model (use model_provider, not provider)
model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

# 2. Define a detailed System Prompt to shape the agent's behavior
system_prompt = """
You are a highly analytical AI Weather Assistant. 
Your tone must be professional, warm, and structured.

RULES:
1. Since you do not have live tools, you cannot provide real-time weather forecasts.
2. If a user asks about current or real-time weather, you must politely inform them that you don't have access to live instruments, but provide historical climate expectations for that region during this time of the year.
3. Always structure your answer into two sections:
   - [STATUS]: Acknowledge your limitations regarding live data.
   - [CLIMATE BRIEF]: Provide typical seasonal weather behavior for the requested city.
"""

# 3. Create the conversation history list
# The SystemMessage MUST always be the very first message in the list
messages = [
    SystemMessage(content=system_prompt),
    HumanMessage(content="What is the weather in Tokyo right now?")
]

# 4. Invoke the model
response = model.invoke(messages)

# 5. Print the final answer
print(response.content)
```

MESSAGE MUST HAVE 

1)  ROLE=identifies the message type (e.g., "system", "user",     "assistant", "tool")

2)  CONTENT=the actual text or data of the message

3)  METADATA=additional information about the message (e.g., timestamp, author, etc.)


```python
# pip install langchain-core
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage

# 1. Creating the messages with ROLE, CONTENT, and METADATA
messages = [
    # ROLE: System
    SystemMessage(
        content="You are a witty, helpful AI coding partner.", # CONTENT
        additional_kwargs={"version": "2.0"}                  # METADATA
    ),
    
    # ROLE: User (called HumanMessage in LangChain)
    HumanMessage(
        content="Explain LangChain messages to me.",          # CONTENT
        additional_kwargs={"user_id": "dev_404"}              # METADATA
    ),
    
    # ROLE: Assistant (called AIMessage in LangChain)
    AIMessage(
        content="LangChain uses explicit objects for roles!",  # CONTENT
        additional_kwargs={"tokens_used": 15}                  # METADATA
    )
]

# 2. How to access the components in your LangChain loop
for msg in messages:
    # LangChain identifies the type using msg.type (returns 'system', 'human', 'ai', etc.)
    print(f"ROLE:      {msg.type.upper()}") 
    print(f"CONTENT:   {msg.content}")
    print(f"METADATA:  {msg.additional_kwargs}")
    print("-" * 40)
```

```python
# CELL 1: Define and create the tool
from langchain_core.tools import tool

@tool
def get_weather_tool(location: str) -> str:
    """Look up the current weather for a specific city."""
    # In real life, this would call a live weather API.
    # For this practice, we will just use a simple if/else statement.
    city = location.lower()
    if "tokyo" in city:
        return "Sunny, 25°C"
    elif "london" in city:
        return "Raining, 12°C"
    else:
        return "Partly Cloudy, 20°C"

print("Tool 'get_weather_tool' has been successfully created and registered!")
```

```python
# CELL 2: Run the tool and save the answer
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage

# 1. Conversation starts
user_msg = HumanMessage(content="What is the weather in Tokyo?")

# 2. AI asks to use the tool (ID is 'call_123')
ai_msg = AIMessage(
    content="", 
    tool_calls=[{"name": "get_weather_tool", "args": {"location": "Tokyo"}, "id": "call_123"}]
)

# 3. Your Python code runs the tool from Cell 1
tool_result = get_weather_tool.invoke({"location": "Tokyo"})

# 4. Save the tool's answer using the matching ID
tool_msg = ToolMessage(content=tool_result, tool_call_id="call_123")


# --- See the result ---
print(user_msg)
print(ai_msg)
print(tool_msg)
```
